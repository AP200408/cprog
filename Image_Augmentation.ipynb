```python
#Question 1

from google.colab import files
from fastai.vision.all import *
import random
import matplotlib.pyplot as plt

uploaded = files.upload()
!pip install fastai --upgrade
filename = next(iter(uploaded))
im_path = Path(filename)
img = PILImage.create(im_path)

# Generate 10 rotated images with random angles between 0 and 80 degrees
for i in range(10):
  angle = random.randint(0, 80)
  rotated_image = img.rotate(angle)
  plt.figure(figsize=(5,5))
  plt.imshow(rotated_image)
  plt.title(f"Rotated by {angle} degrees")
  plt.show()
```

```python
#define cropping box (left,upper,right,lower)
width,height = img.size
left = (width - 200)/2
upper = (height-300)/2
right = (width+200)/2
lower = (height+10)/2
crop_box = (left,upper,right,lower)
#crop the image
cropped_image = img.crop(crop_box)
#display the image
plt.figure(figsize=(5,5))
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.show()
```
